```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = 'data/test_1_2.npy'
x = np.load(filename,allow_pickle=True).item()
x.keys()
```

```python
plt.figure(figsize=(10,10),dpi=120)

plt.subplot(3,3,1)
plt.imshow(np.log10(np.abs(x['stft'][0][:,:,0])))
plt.title('source 1')

plt.subplot(3,3,2)
plt.imshow(np.log10(np.abs(x['stft'][1][:,:,0])))
plt.title('source 2')

plt.subplot(3,3,3)
plt.imshow(np.log10(np.abs(x['mix'][:,:,0])))
plt.title('mix')

plt.subplot(3,3,4)
plt.imshow(np.log10(np.abs(x['mask'][0][:,:,0])))
plt.title('mask 1')

plt.subplot(3,3,5)
plt.imshow(np.log10(np.abs(x['mask'][1][:,:,0])))
plt.title('mask 2');
```
